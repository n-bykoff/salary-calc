```python
from bs4 import BeautifulSoup
import requests
import json
```

```python
response = requests.get('https://www.officeholidays.com/countries/russia/2022')
soup = BeautifulSoup(response.content, 'html.parser')

rows_days_list = soup.find_all('time')
rows_names_list = soup.find_all('a', 'country-listing')
rows_comments_list = soup.find_all('td', 'comments')
```

```python
print(soup.prettify())
```

```python
months_dict = {
    'Jan': 1,
    'Feb': 2,
    'Mar': 3,
    'Apr': 4,
    'May': 5,
    'Jun': 6,
    'Jul': 7,
    'Aug': 8,
    'Sep': 9,
    'Oct': 10,
    'Nov': 11,
    'Dec': 12
}
```

```python
holidays = {}

for row_n in range(0, len(rows_names_list)):
    
    date_str = rows_days_list[row_n].text
    
    month = date_str.split(' ')[0]
    day = int(date_str.split(' ')[1])
    name = rows_names_list[row_n].text
    comment = rows_comments_list[row_n].text
    
    if months_dict[month] not in holidays.keys():
        if comment.lower() != 'not a public holiday':
            holidays[months_dict[month]] = [(day, name, comment)]
    else:
        if comment.lower() != 'not a public holiday':
            month_holidays = holidays[months_dict[month]]
            month_holidays.append((day, name, comment))
            holidays[months_dict[month]] = month_holidays
    
    print(month, day, name, comment)
```

```python
with open('holidays_2022.json', 'w') as f:
    json.dump(holidays, f)
```
